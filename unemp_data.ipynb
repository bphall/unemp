```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', None)

import warnings
warnings.filterwarnings("ignore")
```

```python
raw = pd.read_excel('SeriesReport-20200825152426_c3387f.xlsx')
```

```python
df = raw[11:]
```

```python
df.columns = raw.iloc[10].values
```

```python
unemp_vals = []
for i in range(0, 73):
    unemp_vals.extend(df.drop('Year', axis=1).fillna(10.2).iloc[i].values)
```

```python
years = pd.date_range('1948', '2021', freq='M')
```

```python
time = pd.Series(unemp_vals, years)
```

```python
plt.figure(figsize=(20,12))
time[:].plot(label='umemployment')

plt.axvspan('2001-03-01', '2001-11-01', facecolor='r', alpha=0.05, label='early 2000s recession', color='grey')
plt.axvspan('2007-12-01', '2009-06-01', facecolor='r', alpha=0.05, label='housing crisis', color='blue')
plt.axvspan('2020-04-30', '2020-12-01', facecolor='r', alpha=0.05, label='covid19')

plt.legend()
plt.xlabel('Years', fontsize=20)
# plt.xticks(ticks=np.arange(0, 876,20))
plt.ylabel('Percent U.S. Unemployment', fontsize=20)
plt.title('U.S. Unemployment by Year Since 1948', fontsize=20)
plt.show()
```
